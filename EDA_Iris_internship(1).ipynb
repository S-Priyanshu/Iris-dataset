```python
def graph(y):
	sns.boxplot(x="species", y=y, data=df)

plt.figure(figsize=(10,10))


plt.subplot(221) #Specifies a 2x2 grid of subplots and selects the first subplot (top-left corner).
graph('sepal_length')

plt.subplot(222)  # Select the second subplot (top-right corner)
graph('sepal_width')

plt.subplot(223) # Select the third subplot (bottom-left corner)
graph('petal_length')

plt.subplot(224)   # Select the fourth subplot (bottom-right corner)
graph('petal_width')

plt.show()
```





*   Species Setosa has the smallest features and less distributed with some outliers.
*   Species Versicolor has the average features.

*   Species Virginica has the highest features








# **HEAP MAP**

```python
sns.heatmap(df.corr(), linecolor='white', linewidths=1)
```

A heatmap is a graphical representation of data in a matrix form, where values are represented by colors.



*  df.corr(): Computes the pairwise correlation matrix for numerical columns in the DataFrame.
*   sns.heatmap(): Creates a heatmap using Seaborn, where the correlation values are represented by colors.


*   linecolor='white': Sets the color of the lines between cells in the heatmap to white.
*   linewidths=1: Sets the width of the lines between cells to 1.

The heatmap allows you to visually assess the correlation between different numerical features. The colors and the corresponding values in the cells provide a quick overview of the strength and direction of the relationships between variables. Lighter colors typically indicate stronger correlations, while darker colors indicate weaker correlations.










```python
sns.heatmap(df.corr(),cmap="YlGnBu", linecolor='white', linewidths=1)
```



*   cmap="YlGnBu" parameter to the sns.heatmap() function, which specifies the color map to be used for the heatmap




By specifying the color map, you can control the aesthetic appearance of the heatmap. Different color maps may be more suitable for different types of data or personal preferences. "YlGnBu" (yellow,Green ,Blue)is a popular choice for visualizing correlations as it provides a clear progression from low to high values with distinct color shades.


```python
sns.heatmap(df.corr(), linecolor='white',linewidths=1, annot=True, fmt="f")
plt.show()
```




*   Petal width and petal length have high correlations.
*   Petal length and sepal width have good correlations.
*   
Petal Width and Sepal length have good correlations.



*   annot=True: Enables the annotation of each cell with the numeric value.
*  fmt="f": Specifies the format of the numeric values to be displayed. In this case, "f" stands for float format.
